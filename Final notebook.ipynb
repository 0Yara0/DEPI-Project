```python
df_cleaned['Ship Date']=pd.to_datetime(df_cleaned['Ship Date'], format='%d/%m/%Y')
df_cleaned['Order Date']=pd.to_datetime(df_cleaned['Order Date'], format='%d/%m/%Y')
df_cleaned['Order ID']= df['Order ID'].astype("string")
df_cleaned['Ship Mode']= df['Ship Mode'].astype("string")
df_cleaned['Customer ID']= df['Customer ID'].astype("string")
df_cleaned['Segment']= df['Segment'].astype("string")
df_cleaned['Country']= df['Country'].astype("string")
df_cleaned['City']= df['City'].astype("string")
df_cleaned['State']= df['State'].astype("string")
df_cleaned['Region']= df['Region'].astype("string")
df_cleaned['Product ID']= df['Product ID'].astype("string")
df_cleaned['Category']= df['Category'].astype("string")
df_cleaned['Sub-Category']= df['Sub-Category'].astype("string")
df_cleaned['Product Name']= df['Product Name'].astype("string")
df_cleaned['Category']= df['Category'].astype("string")
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=df_cleaned.index, y='Sales', data=df_cleaned, label='Sales', color='blue')
plt.title('Sales Scatter Plot with Outliers')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
Q1 = df_cleaned['Sales'].quantile(0.25)
Q3 = df_cleaned['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df_cleaned[(df_cleaned['Sales'] < lower_bound) | (df_cleaned['Sales'] > upper_bound)]
non_outliers = df_cleaned[(df_cleaned['Sales'] >= lower_bound) & (df_cleaned['Sales'] <= upper_bound)]


plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(non_outliers.index, non_outliers['Sales'], color='blue', label='Non-Outliers')
plt.scatter(outliers.index, outliers['Sales'], color='red', label='Outliers')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.title('Scatter Plot of Sales Outliers')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
df_cleaned['Order Date'] = pd.to_datetime(df_cleaned['Order Date'], errors='coerce')
```

```python
monthly_sales = df_cleaned.groupby([pd.Grouper(key='Order Date', freq='M'), 'Category'])['Sales'].sum().reset_index()
```

```python
monthly_sales_pivot = monthly_sales.pivot(index='Order Date', columns='Category', values='Sales')
```

```python
monthly_sales_pivot.plot(figsize=(10, 6), title='Monthly Sales by Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
furniture_sales = monthly_sales_pivot['Furniture'].dropna().reset_index()
```

```python
furniture_sales['Month_Num'] = np.arange(len(furniture_sales))
```

```python
X = furniture_sales[['Month_Num']]
y = furniture_sales['Furniture']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(furniture_sales['Order Date'], y, label='Actual Sales', color='blue')
plt.plot(furniture_sales['Order Date'][len(X_train):], y_pred, label='Predicted Sales', color='red')
plt.title('Furniture Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
future_months = np.arange(len(furniture_sales), len(furniture_sales) + 12).reshape(-1, 1)
future_sales_pred = model.predict(future_months)
```

```python
future_dates = pd.date_range(furniture_sales['Order Date'].max(), periods=12, freq='ME')
plt.plot(future_dates, future_sales_pred, label='Forecasted Sales', color='green')
plt.title('Furniture Sales Forecast for 2019')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
for category in monthly_sales_pivot.columns:
    category_sales = monthly_sales_pivot[category].dropna().reset_index()
    category_sales['Month_Num'] = range(len(category_sales))
    
    X = category_sales[['Month_Num']]
    y = category_sales[category]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    plt.figure(figsize=(10, 6))
    plt.plot(category_sales['Order Date'], y, label=f'Actual {category} Sales', color='blue')
    plt.plot(category_sales['Order Date'][len(X_train):], y_pred, label=f'Predicted {category} Sales', color='red')
    plt.title(f'{category} Sales Forecast')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.legend()
    plt.show()
```
